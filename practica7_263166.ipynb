Practica #7. Análisis de un Dataset con Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#import pandas as pd
```

***Carga y exploración del dataset***


```python
### Exploracion de datos usando .head(), .info(), .describe()

df = sns.load_dataset('iris')
#print(df)

print('Primeros 10 renglones del dataset Iris \n',df.head(10), '\n')
print('Resumen del dataset Iris')
print( df.info(), '\n')

print('Resumen estadistico del dataset Iris \n',df.describe(), '\n')

### Revision de valores nulos
valNull = df.isnull().sum()
print('Cantidad de valores nulos por columna \n',valNull)
```

**Creacion de graficas usando Matplotlib**


**a) Gráfico de barras:**

```python
### Agrupacion por especie y calculo de media para  petal_length y petal_width

sortSpecies = df.groupby('species')[['petal_length','petal_width']].mean()
#print(sortSpecies)

plt.figure(figsize=(7,5))

sortSpecies.plot(kind = 'bar', color = ['green','orange'])

plt.title('Comparacion del promedio de ancho y largo de petalo por especie')
plt.xlabel('Especie')
plt.ylabel('Media de anchura y largo')

plt.show()
```

**b) Histograma:**

```python
### Distribucion de datos de sepal_lenght

df_SL = df.dropna(subset = ['sepal_length'])
#print(df_SL)

plt.figure(figsize=(10,8))
plt.hist(df_SL['sepal_length'], bins = 10, color = 'aqua', edgecolor = 'blue')


plt.title('Distribucion del largo de los sepalos')
plt.xlabel('Largo de sepalos')
plt.ylabel('Frecuencia')

plt.show()

```

c) Gráfico de dispersión (scatter plot):

Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
### Relacion del largo del pétalo con el ancho del pétalo

especies = df['species'].unique()
colores = {'setosa' : 'aqua' , 'versicolor' : 'pink' , 'virginica' : 'orange'}

for i in especies:
  subset = df[df['species']== i]
  plt.scatter(subset['petal_width'], subset['petal_length'], label = i, color = colores.get(i,'black'))  #alpha = 0.5, c = df['petal_width'], cmap = 'coolwarm')

plt.title('Relacion entre petal_length y petal_width por especie')
plt.xlabel('Largo de sepalos')
plt.ylabel('Frecuencia')
plt.legend(title = 'Especie')
plt.show()

```
