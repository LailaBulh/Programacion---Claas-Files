### **1) Operaciones básicas**

```python
import numpy as np
import random
```

```python
### Creacion de arreglo
arr = []

for i in range(10):
  arr.append(random.randint(1, 10))

print('Valores del arreglo:\n', arr)


### Cálculo de valores estadísticos
media = np.mean(arr)
print('\nLa media para los numeros del arreglo es:\n', media)

mediana = np.median(arr)
print('La mediana para los numeros del arreglo es:\n', mediana)

dev_std = np.std(arr)
print('La desviación estándar para el arreglo es:\n', dev_std)

arrary2 = np.multiply(arr, 2)
print('\nEl arreglo multiplicado por 2 es:\n', arrary2)
```

### **3) Estadística descriptivo**

```python
import statistics as stat
```

```python
### Creacion de arreglo
arr2 = []

for i in range(100):
  arr2.append(random.randint(0,50))

print('Arreglo de 100 valores entre 0 y 50\n', arr2)

### Calculo valores estadisticos
media2 = np.mean(arr2)
print('\nLa media para los numeros del arreglo es:\n', media2)

mediana2 = np.median(arr2)
print('\nLa mediana para los numeros del arreglo es:\n', mediana2)

moda_arr2 = stat.mode(arr2)
print('\nLa moda para los numeros del arreglo es:\n', moda_arr2)

min_arr2 = min(arr2)
print('\nEl valor mínimo del arreglo es: ', min_arr2)

max_arr2 = max(arr2)
print('\nEl valor máximo del arreglo es: ', max_arr2)
```

### **8) Scater Plot**

```python
from sklearn.datasets import load_iris
```

```python
### Carga dataset 'iris'

df = sns.load_dataset('iris')
print('Vista previa datasest Iris\n', df.head())

especies = df['species'].unique()
print('\nEspecies de flores:\n', especies)

colores = {'setosa' : 'aqua' , 'versicolor' : 'pink' , 'virginica' : 'orange'}

### Scatterplot

for i in especies:
  subset = df[df['species']== i]
  plt.scatter(subset['sepal_width'], subset['sepal_length'], label = i, color = colores.get(i,'black'))

plt.title('Relación entre ancho  y longitud del sépalo')
plt.xlabel('Ancho del sépalo')
plt.ylabel('Longitud del sépalo')
plt.legend()
plt.show()
```

### **10) Gráfico de línea con Media Móvil**

### **11) Datos Temporales con análisis estadístico**

```python
### Declaracion de arreglo de fechas inicializado en 2024-12-01

fechas3 = pd.date_range(start='2024-12-01', periods=200)

df3 = pd.DataFrame({'fecha': fechas3})

### Columna nueva en df para agregar valores
df3['Values'] = [random.randint(1,100) for i in range(len(df3))]
print('\nDataFrame actualizado:\n', df3)

media = df3['Values'].mean()
desviacionStd = df3['Values'].std()

###  Calculo de media y desviacion estandar
print('\nLa media es: ', media)
print('\nLa desviacion estandar es: ', desviacionStd)


### Calculo de bandas de desviacion estandar
bandaSuperior = media + desviacionStd
bandaInferior = media - desviacionStd

print('\nBanda superior: ', bandaSuperior)
print('\nBanda inferior: ', bandaInferior)


###
plt.figure(figsize= (10, 8))

plt.plot(df3['fecha'], df3['Values'], color= 'blue')
plt.axhline(media, color= 'red', linestyle= 'dashed', label= 'media')
plt.axhline(bandaSuperior, color= 'orange', linestyle= 'dashdot', label= 'bandaSuperior')
plt.axhline(bandaInferior, color= 'orange', linestyle= 'dashdot', label= 'bandaInferior')



plt.title('Serie temporal con análisis estadístico')
plt.xlabel('Fecha')
plt.ylabel('Valores y media')
plt.legend()
plt.show()

```
