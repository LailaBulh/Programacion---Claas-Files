***1. Carga y exploración del dataset***


```python
### Exploracion de datos usando .head(), .info(), .describe()

df = sns.load_dataset('tips')
#print(df)

print('Primeros 10 renglones del dataset Tips \n',df.head(10), '\n')
print('Resumen del dataset Tips')
print( df.info(), '\n')

print('Resumen estadistico del dataset Tips \n',df.describe(), '\n')

### Revision de valores nulos
valNull = df.isnull().sum()
print('Cantidad de valores nulos por columna \n',valNull)
```

**a) Mapa de calor**

```python
#total_bill tip. size

df_HM = df[['total_bill','tip', 'size']]
print(df_HM)

sbn.heatmap(df_HM.corr(), cmap = 'mako')

plt.title('Correlacion entre las variables total_bill, tip, size')
plt.xlabel('Size')
plt.ylabel('Nombre de las variables')
```

**b) Diagrama de violin**

```python
### Distribucion de propinas dependiendo el dia de la semana

### dataframe para las columnas de interes tip y day
data = df[['tip','day']]
#print(data)


sbn.set(style = 'whitegrid')
sbn.violinplot(x = 'day', y = 'tip', hue = 'day', data = data, palette= 'Paired')

plt.title('Propinas en cada dia de la semana')
plt.xlabel('Dia de la semana')
plt.ylabel('Propina')

```

**c) Grafico de dispersion**

```python

sns.scatterplot(x = 'total_bill', y = 'tip', data = df, hue = 'day', palette= 'mako')

plt.title('Relacion de propinas por total de cuenta')
plt.xlabel('Total por cuenta')
plt.ylabel('Propina')
```

**d) Grafico de barras**

```python


### Se calculo la media como verificacion personal que se graficaban los datos correctos
avg_tip = df.groupby('sex')['tip'].mean()
#print(avg_tip)

sbn.barplot(x = 'sex', y = 'tip', data = df,  palette = 'muted')


plt.title('Propinas promedio por genero ')
plt.xlabel('Genero')
plt.ylabel('Propina')
```
